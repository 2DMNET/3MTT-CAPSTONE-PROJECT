DETECTING OUTLIERS FROM THE DATASET

```python
# Function to detect outliers using the IQR method
def detect_outliers_iqr(encoded_data, column):
    Q1 = encoded_data[column].quantile(0.25)
    Q3 = encoded_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = encoded_data[(encoded_data[column] < lower_bound) | (encoded_data[column] > upper_bound)]
    return outliers

# Analyze numerical columns for outliers
numerical_columns = encoded_data.select_dtypes(include=['float64', 'int64']).columns

outliers_summary = {}
for col in numerical_columns:
    outliers = detect_outliers_iqr(encoded_data, col)
    outliers_summary[col] = len(outliers)
    print(f"Outliers in {col}: {len(outliers)}")

# Visualize Outliers with Boxplots

plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 5, i)
    sns.boxplot(data=encoded_data, y=col, color='lightblue')
    plt.title(f"Boxplot for {col}")
    plt.tight_layout()
plt.show()
```

TREATING THE OUTLIER USING CAPPING OR FLOOR METHOD

```python
# Cap or floor outliers
def cap_outliers(df, column):
    Q1 = encoded_data[column].quantile(0.25)
    Q3 = encoded_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    encoded_data[column] = np.where(encoded_data[column] < lower_bound, lower_bound, encoded_data[column])
    encoded_data[column] = np.where(df[column] > upper_bound, upper_bound, encoded_data[column])
    return encoded_data

for col in numerical_columns:
    data = cap_outliers(encoded_data, col)

print("Dataset after capping outliers:")
print(data.describe())
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import itertools

# Function to evaluate the ARIMA model with a given set of parameters (p, d, q)
def evaluate_arima_model(X, p, d, q):
    train_size = int(len(X) * 0.8)
    train, test = X[0:train_size], X[train_size:]

    # Fit the ARIMA model
    model = ARIMA(train, order=(p,d,q))
    model_fit = model.fit()

    # Forecast the next 'len(test)' points
    forecast = model_fit.forecast(steps=len(test))

    # Evaluate the performance using MAE
    mae = mean_absolute_error(test, forecast)
    return mae

# Define the parameter grid for ARIMA (p, d, q)
p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)

# Create a list of all parameter combinations
param_grid = list(itertools.product(p_values, d_values, q_values))

# Store results
results = []

# Grid search to find the best combination of parameters
for param in param_grid:
    p, d, q = param
    mae = evaluate_arima_model(data['Confirmed'].values, p, d, q)
    results.append((p, d, q, mae))

# Sort results by MAE
results.sort(key=lambda x: x[3])
best_params = results[0]
print(f"Best ARIMA parameters: p={best_params[0]}, d={best_params[1]}, q={best_params[2]}")
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Scale the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data_normalized['Confirmed'].values.reshape(-1, 1))

# Convert the data to sequences for LSTM (time-step approach)
def create_dataset(data_normalized, time_step=1):
    X, y = [], []
    for i in range(len(data_normalized)-time_step-1):
        X.append(data_normalized[i:(i+time_step), 0])
        y.append(data_normalized[i + time_step, 0])
    return np.array(X), np.array(y)

# Define time_step (number of previous days to predict the next day)
time_step = 10
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be suitable for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Create LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model_lstm.add(LSTM(units=50, return_sequences=False))
model_lstm.add(Dense(units=1))

# Compile and fit the model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')
model_lstm.fit(X, y, epochs=20, batch_size=32)

# Make predictions (forecast the next 14 days)
predicted = model_lstm.predict(X[-14:].reshape(14, time_step, 1))
predicted = scaler.inverse_transform(predicted)

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Confirmed'], label='Historical Data')
plt.plot(pd.date_range(data.index[-1], periods=15, freq='D')[1:], predicted, label='Forecast (LSTM)', color='red')
plt.title('COVID-19 Forecast using LSTM (Next 14 Days)')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.show()
```

```python
# Evaluate LSTM predictions
# Actual values for the forecasted period (e.g., next 14 days)
actual_lstm = data_normalized['Confirmed'][-14:].values
predicted_lstm = predicted.flatten()  # LSTM forecast

# Call the evaluation function
evaluate_model(actual_lstm, predicted_lstm)
```
